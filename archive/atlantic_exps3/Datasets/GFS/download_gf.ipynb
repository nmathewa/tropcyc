```python

file_name= 'gfs.0p25.2015011500.f000.grib2'
date_txt = datetime.datetime(2015,1,15,0)
new_date = date_txt.strftime('%Y%m%d%H')

dates = pd.date_range(start='2019-08-01',end='2022-12-31',freq='6H')


dft_dates = pd.DataFrame(dates,columns=['date'])

dft_dates['date_txt'] = dft_dates['date'].dt.strftime('%Y%m%d%H')

dft_dates['days'] = abs(((dft_dates['date'].iloc[0] - dft_dates['date']).dt.days).astype(int))

# for days fill with np.arange(0,48,6)

forecast_id= []

for jj in range(int(len(dft_dates))):
    for ii in range(0,48,6):
        forecast_id.append(ii)


dft_dates['forecast_id'] = forecast_id[:len(dft_dates)]

dft_dates
```

```python


dft_dates['filename'] = 'gfs.0p25.' + dft_dates['date_txt'] + '.f' + dft_dates['forecast_id'].astype(str).str.zfill(3) + '.grib2'

dft_dates.sort_values(by=['date','forecast_id'],inplace=True)

dft_dates
```

```python
import sys, os
from urllib.request import build_opener

opener = build_opener()

filelist = [
  url
]

for file in filelist:
    ofile = os.path.basename(file)
    sys.stdout.write("downloading " + ofile + " ... ")
    sys.stdout.flush()
    infile = opener.open(file)
    outfile = open(ofile, "wb")
    outfile.write(infile.read())
    outfile.close()
    sys.stdout.write("done\n")
```
